```python
import pandas as pd
import numpy as np
```

```python
# Import Data

df = pd.read_csv('Data/Basic_Stats.csv')
```

```python
csp = pd.read_csv('Data/Career_Stats_Passing.csv')
csp.head()
```

```python
csp.info()
```

```python
csp.replace('--', 0, inplace=True)

csp['Passing Yards'].replace(',', '', inplace=True, regex=True)
csp['Longest Pass'].replace('T', '', inplace=True, regex=True)

csp = csp.astype({'Passes Attempted': 'int64', 
                 'Passes Completed': 'int64', 
                 'Completion Percentage': 'float64',
                 'Pass Attempts Per Game': 'float64', 
                 'Passing Yards': 'int64',
                 'Passing Yards Per Attempt': 'float64', 
                 'Passing Yards Per Game': 'float64',
                 'TD Passes': 'int64',
                 'Percentage of TDs per Attempts': 'float64',
                 'Ints': 'int64',
                 'Int Rate': 'float64',
                 'Longest Pass': 'int64',
                 'Passes Longer than 20 Yards': 'int64',
                 'Passes Longer than 40 Yards': 'int64',
                  'Sacks': 'int64',
                 'Sacked Yards Lost': 'int64',
                 'Passer Rating': 'int64'
                })

csp.head()
```

```python
csp.info()
```

```python
csp_grouped = csp.groupby('Player Id').agg(start_year = ('Year', 'min'),
                            end_year = ('Year', 'max'),
                            team_1 = ('Team', lambda x: x.value_counts().index[0]),
                            team_2 = ('Team', lambda x: x.value_counts().index[1] if len(x.value_counts()) > 1 else np.NaN), 
                            team_3 = ('Team', lambda x: x.value_counts().index[2] if len(x.value_counts()) > 2 else np.NaN),
                            pass_att = ('Passes Attempted', 'sum'),
                            pass_compl = ('Passes Completed', 'sum'),
                            comp_perc = ('Completion Percentage', 'mean'),
                            pass_att_per_gm = ('Pass Attempts Per Game', 'mean'),
                            pass_yds = ('Passing Yards', 'sum'),
                            pass_yds_per_att = ('Passing Yards Per Attempt', 'mean'),
                            pass_yds_per_gm = ('Passing Yards Per Game', 'sum'),
                            pass_tds = ('TD Passes', 'sum'),
                            td_per_att = ('Percentage of TDs per Attempts', 'sum'),
                            ints = ('Ints', 'sum'),
                            int_rate = ('Int Rate', 'mean'),
                            longest_pass = ('Longest Pass', 'max'),
                            pass_20yds = ('Passes Longer than 20 Yards', 'sum'),               
                            pass_40yds = ('Passes Longer than 40 Yards', 'sum'),               
                            sacks = ('Sacks', 'sum'),               
                            yds_lost = ('Sacked Yards Lost', 'sum'),
                            Passer_rating = ('Passer Rating', 'mean')               
                            )
```

```python
csp_grouped
```

```python
df = df.merge(csp_grouped, how='outer', on='Player Id')

df.head()
```

```python
csr = pd.read_csv('Data/Career_Stats_Receiving.csv')

csr.info()
```

```python
csr.drop(columns=['Longest Reception'], inplace=True, errors='ignore')

csr.replace('--', 0, inplace=True)
csr['Receiving Yards'].replace(',', '', inplace=True, regex=True)

csr = csr.astype({'Receptions': 'int64', 
            'Receiving Yards': 'int64', 
            'Receiving TDs': 'int64',
            'Receptions Longer than 20 Yards': 'int64', 
            'Receptions Longer than 40 Yards': 'int64',
            'First Down Receptions': 'int64', 
            'Fumbles': 'int64',
            'Yards Per Reception': 'float64',
            'Yards Per Game': 'float64'})
```

```python
csr_grouped = csr.groupby('Player Id').agg(start_year = ('Year', 'min'),
                            end_year = ('Year', 'max'),
                            Team_1 = ('Team', lambda x: x.value_counts().index[0]),
                            Team_2 = ('Team', lambda x: x.value_counts().index[1] if len(x.value_counts()) > 1 else np.NaN), 
                            Team_3 = ('Team', lambda x: x.value_counts().index[2] if len(x.value_counts()) > 2 else np.NaN),
                            games_played = ('Games Played', 'sum'),
                            receptions = ('Receptions', 'sum'),
                            receiving_yards = ('Receiving Yards', 'sum'),
                            receiving_TDs = ('Receiving TDs', 'sum'),
                            receptions_20yds = ('Receptions Longer than 20 Yards', 'sum'),
                            receptions_40yds = ('Receptions Longer than 40 Yards', 'sum'),
                            first_down_recep = ('First Down Receptions', 'sum'),
                            fumbles = ('Fumbles', 'sum'),
                            yds_per_rec = ('Yards Per Reception', 'mean'),
                            yds_per_gm = ('Yards Per Game', 'mean')
                            )
```

```python
df = df.combine_first(csr_grouped.loc[:,'start_year':'games_played'])
```

```python
df = df.merge(csr_grouped.loc[:,'receptions':], how='outer', on='Player Id')
```

```python
csru_grouped = csru.groupby('Player Id').agg(start_year = ('Year', 'min'),
                            end_year = ('Year', 'max'),
                            Team_1 = ('Team', lambda x: x.value_counts().index[0]),
                            Team_2 = ('Team', lambda x: x.value_counts().index[1] if len(x.value_counts()) > 1 else np.NaN), 
                            Team_3 = ('Team', lambda x: x.value_counts().index[2] if len(x.value_counts()) > 2 else np.NaN),
                            games_played = ('Games Played', 'sum'),
                            rush_att = ('Rushing Attempts', 'sum'),
                            rush_att_per_gm = ('Rushing Attempts Per Game', 'mean'),
                            rush_yds = ('Rushing Yards', 'sum'),
                            yds_per_carry = ('Yards Per Carry', 'mean'),
                            rush_yds_per_gm = ('Rushing Yards Per Game', 'mean'),
                            rush_TDs = ('Rushing TDs', 'sum'),
                            first_down_rush = ('Rushing First Downs', 'sum'),
                            first_down_run_perc = ('Percentage of Rushing First Downs', 'mean'),
                            rush_20yds = ('Rushing More Than 20 Yards', 'sum'),
                            rush_40yds = ('Rushing More Than 40 Yards', 'sum'),
                            fumbles = ('Fumbles', 'sum'),
                            )
```

```python
df = df.combine_first(csru_grouped.loc[:,'start_year':'games_played'])
```

```python
df = df.merge(csru_grouped.loc[:,'rush_att':], how='outer', on='Player Id')
```

```python
hof = pd.read_csv('Data/Hall of Fame List.csv')

hof
```

```python
hof.drop(columns = ['indct', 'from', 'to', '1_allpro', 'probowl', 'years_started', 'carav', 'games_played'], inplace=True, errors='ignore')

hof = hof[hof['pos'].isin(['WR', 'RB', 'E', 'FB', 'FL', 'HB', 'TB', 'TE'])]

hof['hof'] = 1

hof
                      
```

```python
split = hof['name'].str.split(' ')
hof['last'] = split.str[1]
hof['first'] = split.str[0]
```

```python
hof['name'] = hof['last'] + str(', ') + hof['first']
hof.drop(columns=['pos', 'last', 'first'], inplace=True)
```

```python
hof[hof['name'] == 'Sharpe, Shannon']
```

```python
df = pd.merge(df, hof, how='outer', left_on='Name', right_on='name')
```
